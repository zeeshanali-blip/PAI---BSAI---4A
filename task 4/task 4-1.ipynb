```python
def is_safe(board, row, col, n):
    
    for i in range(col):
        if board[row][i] == 1:
            return False

    i = row
    j = col
    while i >= 0 and j >= 0:
        if board[i][j] == 1:
            return False
        i -= 1
        j -= 1

    i = row
    j = col
    while i < n and j >= 0:
        if board[i][j] == 1:
            return False
        i += 1
        j -= 1

    return True
```

```python
def solve_nq(board, col, n):
    
    if col >= n:
        return True

    for i in range(n):

        if is_safe(board, i, col, n):

            board[i][col] = 1

            if solve_nq(board, col + 1, n):
                return True

            board[i][col] = 0

    return False
```

```python
def n_queen(n):
    
    board = [[0]*n for _ in range(n)]

    if not solve_nq(board, 0, n):
        print("Solution does not exist")
        return

    for row in board:
        print(row)


n = int(input("Enter number of queens: "))
n_queen(n)
```
