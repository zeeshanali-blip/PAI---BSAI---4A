```python
from collections import deque

def water_jug_bfs():
    visited = set()
    queue = deque([(0, 0)])

    while queue:
        x, y = queue.popleft()

        if (x, y) in visited:
            continue

        visited.add((x, y))

        if x == 2 or y == 2:
            print("Goal Reached:", (x, y))
            return

        possible_states = [
            (4, y),       
            (x, 3),       
            (0, y),       
            (x, 0),        
            (x - min(x, 3-y), y + min(x, 3-y)),  
            (x + min(y, 4-x), y - min(y, 4-x))   
        ]

        for state in possible_states:
            queue.append(state)

water_jug_bfs()
```
