```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\YOGA\Desktop\4th sem\PAI-LAB---BSAI-4A\task 5\image by zeeshan .jpg.jpeg")

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
blur = cv2.GaussianBlur(img, (5,5), 0)

plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))
plt.title("Blur Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
(h, w) = img.shape[:2]
center = (w//2, h//2)

matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, matrix, (w, h))

plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

```python
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap="gray")
plt.title("Binary Image")
plt.axis("off")
plt.show()
```

```python
hist = cv2.equalizeHist(gray)

plt.imshow(hist, cmap="gray")
plt.title("Histogram Equalization")
plt.axis("off")
plt.show()
```

```python
kernel = np.ones((5,5), np.uint8)
erosion = cv2.erode(gray, kernel, iterations=1)

plt.imshow(erosion, cmap="gray")
plt.title("Erosion")
plt.axis("off")
plt.show()
```

```python
denoise = cv2.fastNlMeansDenoisingColored(img, None, 10,10,7,21)

plt.imshow(cv2.cvtColor(denoise, cv2.COLOR_BGR2RGB))
plt.title("Denoised Image")
plt.axis("off")
plt.show()
```
