```python
def greet(name):
    return "Hello, " + name

print(greet("Alice"))

#function 
def add(a, b):
    return a + b

print(add(5, 3))

#control structure
x = 10

if x > 5:
    print("Greater than 5")
else:
    print("Less than or equal to 5")
    
#loops 
for i in range(5):
    print(i)
    
#list
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
print(fruits)

#tuple 
coordinates = (10, 20)
print(coordinates[0])

#sets
unique_numbers = {1, 2, 3, 3, 4}
print(unique_numbers)

#dictionary
student  = {"name": "Alice", "age": 30}
print(student["name"])
```
